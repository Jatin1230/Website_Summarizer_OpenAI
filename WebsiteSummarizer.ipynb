```python
#we have imported all the necessary libraries
import os
import requests
from dotenv import load_dotenv
from bs4 import BeautifulSoup
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
openai = OpenAI()
```

```python
message = "Hey, ChatGPT this side Jatin, how r u?"
response = openai.chat.completions.create(model="gpt-4o-mini", messages=[{"role":"user", "content": message}])
print(response.choices[0].message.content)
```

```python
headers = {
 "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}

class Website:

    def __init__(self, url):
        """
        Create this Website object from the given url using the BeautifulSoup library
        """
        self.url = url
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        self.title = soup.title.string if soup.title else "No title found"
        for irrelevant in soup.body(["script", "style", "img", "input"]):
            irrelevant.decompose()
        self.text = soup.body.get_text(separator="\n", strip=True)
```

```python
test = Website("https://www.w3schools.com/python/python_classes.asp")
print(test.title)
print(test.text)
```

```python
system_prompt ="You are an assitant that extracts and analyses the information from a website\
and provides a short and brief summary, ignore text that might be direction or navigation related\
Respond in markdown."
```

```python
def user_prompt_for(website):
    user_prompt = f"You are viewing a website named {website.title}"
    user_prompt += "\n Here is a short overview of this website; \
    please provide a short and accurate summary of this website in markdown. "
    user_prompt += website.text
    return user_prompt
```

```python
print(user_prompt_for(test))
```

```python
messages = [
    {"role": "system", "content": "You are a foolish assistant"},
    {"role": "user", "content": "What is 2 + 2?"}
]
```

```python
response = openai.chat.completions.create(model="gpt-4o-mini", messages=messages)
print(response.choices[0].message.content)
```

```python
def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_for(website)}
    ]
```

```python
messages_for(test)
```

```python
def summarize(url):
    website = Website(url)
    response = openai.chat.completions.create(
        model = "gpt-4o-mini",
        messages = messages_for(website)
    )
    return response.choices[0].message.content
```

```python
summarize("https://github.com/Jatin1230")
```

```python
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
```

```python
display_summary("https://github.com/Jatin1230")
```
